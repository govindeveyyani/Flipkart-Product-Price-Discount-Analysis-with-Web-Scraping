```python
Phone_name = []
Price = []
Discount = []
Ratings = []

page_num = 260
for i in range(1, int(page_num)+1):
    url = "https://www.flipkart.com/search?q=mobile&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_3_na_na_na&as-pos=2&as-type=RECENT&suggestionId=mobile%7CMobiles&requestId=553a951a-5abd-4cf7-99e5-c08542285669&as-searchtext=mobile&p%5B%5D=facets.rating%255B%255D%3D3%25E2%2598%2585%2B%2526%2Babove&p%5B%5D=facets.discount_range_v1%255B%255D%3D10%2525%2Bor%2Bmore&page="+str(i)
    req = requests.get(url)
    content = BeautifulSoup(req.content, 'html.parser')
    name = content.find_all("div", class_ = "KzDlHZ")
    price = content.find_all("div", class_ = "Nx9bqj _4b5DiR")
    discount = content.find_all("div", class_ = "UkUFwK")
    rating = content.find_all("div", class_ = "XQDdHH")
    print("Phone in page " +str(i))
    print(len(name))

    for i in name:
        Phone_name.append(i.text)
    for i in price:
        Price.append(i.text)
  
    for i in discount:
        Discount.append(i.text)
    for i in rating:
        Ratings.append(i.text)
    
```

```python
df = pd.DataFrame(list(zip(Phone_name, Price, Discount, Ratings)), columns=["product_name", "Price", "Discount", "Ratings", ])
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Price'] = df['Price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
df['Discount_Percentage'] = df['Discount'].str.replace('% off', '').astype(float)

df.drop(columns=['Discount'], inplace=True)


```

```python

df['Brand'] = df['product_name'].apply(lambda x: x.split()[0])




```
